According to UCAR Center for Science Education:
"In this area, known as Tornado Alley, storms are caused when dry cold air moving south from Canada meets warm moist air traveling north from the Gulf of Mexico. Tornadoes can form at any time of year, but most occur in the spring and summer months along with thunderstorms. May and June are usually the peak months for tornadoes."[1]

### Analysis of Tornadoes, Fatalities, and Injuries between 1950 and 2018: 

Due to the numpy.random function used to help generate the data set, the above numbers and graphs do not correspond to the first generated data set from the top of this page, as each time you run a random function, the numbers change.

My analysis is of the data set and graphs immediately above this markdown cell.

One thing I can point out right away from the graph is that injuries from tornadoes almost always outnumber the number of tornadoes themselves and fatalities caused by tornadoes. There are only seven counties out of 105 where this wasn't the case: Cherokee, Douglas, Elk, Franklin, Grant, Jewel, and Osborne. One could infer that maybe there is a better warning system in place for these counties or that the residents there have much better shelter from tornadoes like basements or bunkers. Perhaps there are also fewer people living in these counties and expect fewer tornadoes per year, because the number of tornadoes is also fairly low in nearly all seven counties.

There are some interesting cases like in Ellsworth County where there were only 2 tornadoes, but they must have been terribly destructive, since 57 people died and 464 were injured. I can only imagine that these tornadoes occured in a highly populated area in which the residents did not have proper access to adequate shelter.
In Elm county, mentioned above, there were 28 tornadoes that caused 19 deaths and only 7 injuries.
I would guess that most people in that county had shelter, but if you were caught out, odds of survival were pretty low.

The Grant County numbers are odd and argubly show a flaw with randomly generated (simulated) data like this.
The numbers are showing 0 tornadoes, yet there were 74 deaths and 79 people injured. Given that the above data didn't factor in any other variables which could have caused death and injury, these numbers don't add up at all. Perhaps I should have set the least amount of any of my variables to be 1 instead of 0. Still, even if there were only one tornado, 74 deaths and 79 injuries would be a lot.

### Conclusions:
  
Before generating the data set, I had assumed that more tornadoes equals more fatalities and injuries. However, after looking at the data, I can confidently conclude that number of tornadoes does not always correlate to death and injuries quite the way one would expect. There are many other factors to consider that could cause this: type of shelter, population density in the area, strength of the tornado, etc. 
I can also conclude that simulated data doesn't always make sense in the real world like in the Grant County example where 0 tornadoes caused 74 deaths and 79 injuries.
