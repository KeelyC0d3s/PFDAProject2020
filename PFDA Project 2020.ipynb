Introduction: 

For this project, I decided to venture into the natural real-world phenomenon of weather and see what kind of data was out there. I eventually decided on tornadoes in Kansas. 

Kansas is a State in the Midwestern United States situated in a place commonly known as 'Tornado Alley' due to the frequency of occurances of tornadoes.

According to UCAR Center for Science Education:
"In this area, known as Tornado Alley, storms are caused when dry cold air moving south from Canada meets warm moist air traveling north from the Gulf of Mexico. Tornadoes can form at any time of year, but most occur in the spring and summer months along with thunderstorms. May and June are usually the peak months for tornadoes."[1]

I found a dataset at https://www.weather.gov/media/top/SWAW_2019.pdf that had a chart of tornadoes by county in the State of Kansas between the years of 1950 and 2018. I decided to use the same variables in this chart for my simulated data set.
